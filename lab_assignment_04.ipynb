```python
import pandas as pd
#read file
df = pd.read_csv (r'/content/lab-assignment-04-student-performance.csv')
df.head()

```

```python
#(1) Description of all the columns: for each colmun, count number of rows, number of unqiue element, the frequency of each unqiue element.
#number of rows
rows = len(df.axes[0])
print("Number of Rows: ", rows)
#number of unique elements in column
n = df.nunique(axis=0)
print("No.of.unique values in each column :\n",
      n)
# Frequency of unique element
freq = df.stack(dropna=False)
val_uniq=freq.groupby(level=0).apply(lambda x: x[~x.duplicated()]).value_counts(dropna=False)
print("Frequency of unique values :\n",
      val_uniq)
```

```python
#(2) Visualize the count for each value in the 'Study Hours' column. Note: Study Hours values comes from this >= 6 hours Good >= 4 hours Average < 2 hours Poor.
study_hours = pd.value_counts(df['sh']).plot.bar()
```

```python
#3) Using factorplot to show the relationship between numerical and categorical value. 
#Please use figure to show relation between [Study Hours] and [End Semester Percentage]. Percentage is calculated based on the following:
#if percentage >=80 then Best 
#If percentage >= 60 but less than 80 then Very Good 
#If percentage >= 45 but less than 60 then Good 
#If Percentage >= 30 but less than 45 then Pass If Percentage < 30 then Fail
import seaborn as sns
import matplotlib.pyplot as plt
#change categorical end of semester percetage to numeric
df['esp'].replace(['Best', 'Vg', 'Good', 'Pass'],
                        [0, 1, 2, 3], inplace=True)

# point plot(by default)
sns.factorplot(x ='sh', y ='esp', data = df)
 
plt.legend(['0=Best', '1=Very Good', '2=Good', '3=Pass'])

```

```python
#Get the correlation from the data, then plug it to heatmap function and show it as following:
#change categorical to numerical
df['ge'].replace(['F','M'],[1, 2], inplace=True)
df['cst'].replace(['G','OBC','MOBC','SC','ST'],[1, 2, 3, 4, 5], inplace=True)
df['tnp'].replace(['Best', 'Vg', 'Good', 'Pass'],[1, 2, 3, 4], inplace=True)
df['twp'].replace(['Best', 'Vg', 'Good', 'Pass'],[1, 2, 3, 4], inplace=True)
df['iap'].replace(['Best', 'Vg', 'Good', 'Pass'],[1, 2, 3, 4], inplace=True)
df['arr'].replace(['Y','N'],[1, 2], inplace=True)
df['ms'].replace(['Unmarried'],[1], inplace=True)
df['ls'].replace(['T','V'],[1,2], inplace=True)
df['as'].replace(['Paid','Free'],[1,2], inplace=True)
df['fmi'].replace(['Am','High','Low','Medium','Vh'],[1,2,3,4,5], inplace=True)
df['fs'].replace(['Average','Large','Small'],[1,2,3], inplace=True)
df['fq'].replace(['Um','12','10','Il','Degree','Pg',],[1,2,3,4,5,6], inplace=True)
df['mq'].replace(['Um','12','10','Il','Degree','Pg',],[1,2,3,4,5,6], inplace=True)
df['fo'].replace(['Business','Farmer','Others','Retired','Service'],[1,2,3,4,5], inplace=True)
df['mo'].replace(['Business','Housewife','Others','Retired','Service'],[1,2,3,4,5], inplace=True)
df['nf'].replace(['Average','Large','Small'],[1,2,3], inplace=True)
df['sh'].replace(['Average','Good','Poor'],[1,2,3], inplace=True)
df['ss'].replace(['Govt','Private'],[1,2], inplace=True)
df['me'].replace(['Asm','Ben','Eng','Hin'],[1,2,3,4], inplace=True)
df['tt'].replace(['Average','Large','Small'],[1,2,3], inplace=True)
df['atd'].replace(['Average','Good','Poor'],[1,2,3], inplace=True)


#sns.heatmap(df, cmap ='RdYlGn',
sns.heatmap(df, cmap ='RdYlGn', linewidths = 0.30, annot = True)
```
